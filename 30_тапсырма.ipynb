# Деректер құрылымдарының әдістері

**Мақсаты:** Python тіліндегі Тізімдер, Кортеждер, Жиындар және Сөздіктердің барлық негізгі функцияларын практикада қолданып үйрену.


## 1-бөлім. Тізімдер (Lists)

**1-тапсырма: Тізім құру және элемент алу**
`fruits` атты тізім құрыңыз ("apple", "banana", "cherry"). Тізімнің екінші элементін экранға шығарыңыз.


```
# Тізімді құру және екінші элементті шығару
fruits = ["apple", "banana", "cherry"]
print(fruits[1])
```


**2-тапсырма: append()**
Тізімнің соңына "orange" элементін қосыңыз.


```
# Тізім соңына элемент қосу
fruits.append("orange")
print(fruits)
```


**3-тапсырма: insert()**
Тізімнің бірінші позициясына (индексі 0) "lemon" элементін кірістіріңіз.


```
# Белгілі бір орынға элемент кірістіру
fruits.insert(0, "lemon")
print(fruits)
```


**4-тапсырма: extend()**
Басқа тізімді (`more_fruits = ["mango", "grape"]`) негізгі тізімге жалғаңыз.


```
# Екі тізімді біріктіру (жалғау)
more_fruits = ["mango", "grape"]
fruits.extend(more_fruits)
print(fruits)
```


**5-тапсырма: remove()**
Тізімнен "banana" элементін мәні бойынша өшіріңіз.


```
# Элементті мәні бойынша өшіру
fruits.remove("banana")
print(fruits)
```


**6-тапсырма: pop()**
Тізімнің соңғы элементін өшіріп, оны экранға шығарыңыз.


```
# Соңғы элементті алып тастау және көрсету
fruits.pop()
print(fruits)
```


**7-тапсырма: index()**
"cherry" элементінің индексін (реттік нөмірін) табыңыз.


```
# Элементтің индексін табу
fruits.index("cherry")
print(fruits)
```



**8-тапсырма: count()**
Тізімде "apple" сөзі қанша рет кездесетін санаңыз.


```
# Элементтің санын есептеу
fruits.count("apple")
print(fruits)
```


**9-тапсырма: sort() және reverse()**
Сандар тізімін құрыңыз (`numbers = [5, 2, 9, 1]`). Оны алдымен өсу ретімен сұрыптаңыз, содан кейін кері ретпен (reverse) өзгертіңіз.


```
# Тізімді сұрыптау және кері айналдыру
numbers = [5, 2, 9, 1]
numbers.sort()
print(numbers)
numbers.reverse()
print(numbers)
```


## 2-бөлім. Кортеждер (Tuples)

**10-тапсырма: Кортеж құру**
`colors` атты кортеж жасаңыз ("red", "green", "blue").


```
# Кортеж құру
colors = ("red", "green", "blue")
print(colors)
```


**11-тапсырма: Индекс арқылы қатынасу**
Кортеждің соңғы элементін теріс индекс (-1) арқылы шығарыңыз.


```
# Соңғы элементті шығару
print(colors[-1])
```



**12-тапсырма: count()**
Кортежде "red" түсі қанша рет бар екенін анықтаңыз.


```
# Элементтің қайталану санын табу
print(colors.count("red"))
```


**13-тапсырма: index()**
"green" элементінің индексін табыңыз.


```
# Элементтің орналасқан жерін табу
print(colors.index("green"))
```


**14-тапсырма: Unpacking (Айнымалыларға бөлу)**
`coordinates = (10, 20)` кортежін `x` және `y` айнымалыларына бөліп жазыңыз.


```
# Кортежді айнымалыларға тарату (unpacking)
coordinates = (10, 20)
x, y = coordinates
print("x =", x)
print("y =", y)
```


## 3-бөлім. Жиындар (Sets)

**15-тапсырма: Жиын құру**
`my_set` атты жиын жасаңыз (1, 2, 3, 3, 2). Нәтижені шығарып, қайталанатын сандардың жойылғанын тексеріңіз.


```
# Жиын құру және бірегейлікті тексеру
my_set = {1, 2, 3, 3, 2}
print(my_set)
```


**16-тапсырма: add()**
Жиынға 4 санын қосыңыз.


```
# Жиынға жаңа элемент қосу
my_set.add(4)
print(my_set)
```


**17-тапсырма: update()**
Жиынға тізімнен бірнеше элемент қосыңыз (`[5, 6]`).


```
# Жиынды жаңарту (бірнеше элемент қосу)
my_set.update([5, 6])
print(my_set)
```



**18-тапсырма: remove() vs discard()**
`remove()` көмегімен 2 санын, ал `discard()` көмегімен 10 санын (жоқ болса да қате шықпауы үшін) өшіріңіз.


```
# Элементтерді қауіпсіз және қауіпті жолмен өшіру
my_set.remove(2)
my_set.discard(10)
print(my_set)
```


**19-тапсырма: union()**
Екі жиынның бірігуін (барлық бірегей элементтерді) табыңыз: `A = {1, 2}`, `B = {2, 3}`.



```
# Екі жиынның бірігуі (union)
A = {1, 2}
B = {2, 3}
print(A.union(B))
```



**20-тапсырма: intersection()**
Екі жиынның қиылысуын (тек ортақ элементтерді) табыңыз.


```
# Ортақ элементтерді табу (intersection)
print(A.intersection(B))
```


**21-тапсырма: difference()**
`A` жиынында бар, бірақ `B` жиынында жоқ элементтерді табыңыз (айырма).


```
# Жиындардың айырмасын табу
print(A.difference(B))
print(B.difference(A))
```



## 4-бөлім. Сөздіктер (Dictionaries)

**22-тапсырма: Сөздік құру**
`car` атты сөздік жасаңыз ("brand": "Toyota", "model": "Camry", "year": 2020).


```
# Сөздік құру
car = {
  "brand": "Toyota",
  "model": "Camry",
  "year": 2020
}
```


**23-тапсырма: Мәнді алу**
Кілт арқылы "brand" мәнін экранға шығарыңыз.


```
# Кілт бойынша мәнді алу
print(car["brand"])
```



**24-тапсырма: get()**
`get()` әдісін қолданып "color" кілтін сұраңыз. Егер ол жоқ болса, "White" деп қайтарсын.


```
# get әдісін қолдану
print(car.get("color", "White"))
```



**25-тапсырма: Жаңарту**
"year" кілтінің мәнін 2023-ке өзгертіңіз.


```
# Сөздіктегі мәнді өзгерту
car["year"] = 2023
print(car)
```



**26-тапсырма: Жаңа жұп қосу**
Сөздікке "color": "Black" жұбын қосыңыз.


```
# Жаңа кілт-мән жұбын қосу
car["color"] = "Black"
print(car)
```



**27-тапсырма: keys()**
Сөздіктегі барлық кілттерді тізім ретінде шығарыңыз.


```
# Барлық кілттерді алу
print(car.keys())
```



**28-тапсырма: values()**
Сөздіктегі барлық мәндерді тізім ретінде шығарыңыз.


```
# Барлық мәндерді алу
print(car.values())
```


**29-тапсырма: items()**
Сөздіктегі барлық кілт-мән жұптарын кортеж түрінде шығарыңыз.


```
# Барлық жұптарды алу
print(car.items())
```



**30-тапсырма: pop()**
"model" кілтін сөздіктен өшіріп, оның мәнін экранға шығарыңыз.


```
# Кілтті өшіру және мәнін қайтару
print(car.pop("model"))
```
